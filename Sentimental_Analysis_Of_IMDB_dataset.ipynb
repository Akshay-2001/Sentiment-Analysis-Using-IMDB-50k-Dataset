```python
#replacing target column into numerical format i.e., +ve -> 1 and -ve -> 2

df.sentiment.replace("positive", 1, inplace=True)
df.sentiment.replace("negative", 2, inplace=True)
```

```python
df.head()
```

```python
# Text Processing

def data_processing(text):
    text = text.lower()
    text = re.sub('<br />', '', text)                                           # Removing Break Tags
    text = re.sub(r"https\S+|www\S+|http\S+", '', text, flags = re.MULTILINE)   # Removing URLs
    text = re.sub(r'\@w+|\#', '', text)                                         # Removing #s and @
    text = re.sub(r'[^\w\s]', '', text)                                         # Removing Punctuations
    text_tokens = word_tokenize(text)                                           
    filtered_text = [w for w in text_tokens if not w in stop_words]             # Removing Stopwords
    return " ".join(filtered_text)
```

```python
nltk.download('punkt')
```

```python
df.review = df['review'].apply(data_processing)
```

```python
# Checking for Duplicates

duplicated_count = df.duplicated().sum()
print("Number of duplicate entries : ", duplicated_count)
```

```python
# Dropping Duplicate data

df = df.drop_duplicates('review')
```

```python
# Performing Stemming 

stemmer = PorterStemmer()
def stemming(data):
    text = [stemmer.stem(word) for word in data]
    return data
```

```python
df.review = df['review'].apply(lambda x: stemming(x))
```

```python
X = df['review']
Y = df['sentiment']
```

```python
# Vectorizing data 
vect = TfidfVectorizer()
X = vect.fit_transform(df['review'])
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
print("Size of x_train : ", (x_train.shape))
print("Size of x_test : ", (x_test.shape))
print("Size of y_train : ", (y_train.shape))
print("Size of y_test : ", (y_test.shape))
```

```python
#importing libraries for creation of ML model
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import warnings
warnings.filterwarnings('ignore')
```

```python
#Trainig Data with Logistic Regression
logreg = LogisticRegression()
logreg.fit(x_train, y_train)
logreg_pred = logreg.predict(x_test)
logreg_acc = accuracy_score(logreg_pred, y_test)
print("Test Accuracy : {:.2f}%".format(logreg_acc*100))
```

```python
#printing Confusion matrix and classification report for logistic regression
print(confusion_matrix(y_test, logreg_pred))
print("\n")
print(classification_report(y_test, logreg_pred))
```

```python
#Training data on a Multinomial Naive Bayes model
mnb = MultinomialNB()
mnb.fit(x_train, y_train)
mnb_pred = mnb.predict(x_test)
mnb_acc = accuracy_score(mnb_pred, y_test)
print("Test Accuracy : {:.2f}%".format(mnb_acc*100))
```

```python
#printing Confusion matrix and classification report for MultinomialNB model
print(confusion_matrix(y_test, mnb_pred))
print("\n")
print(classification_report(y_test, mnb_pred))
```

```python
#training the data on a Support Vector Classifier Model
svc = LinearSVC()
svc.fit(x_train, y_train)
svc_pred = svc.predict(x_test)
svc_acc = accuracy_score(svc_pred, y_test)
print("Test Accuracy : {:.2f}%".format(svc_acc*100))
```

```python
#printing Confusion matrix and classification report for SVC
print(confusion_matrix(y_test, svc_pred))
print("\n")
print(classification_report(y_test, svc_pred))
```

```python
#trying to increase accuraccy by performing hyper parameter tuning on the svc

from sklearn.model_selection import GridSearchCV
param_grid = {'C':[0.1, 1, 10, 100], 'loss':['hinge', 'squared_hinge']}
grid = GridSearchCV(svc, param_grid, refit=True, verbose = 3)
grid.fit(x_train, y_train)
```

```python
#Printing the best cross violation and best parameter
print("Best cross validation score: {:.2f}".format(grid.best_score_))
print("Best Parameter: ", grid.best_params_)
```

```python
# Applying obtained best parameter to SVC Model
svc = LinearSVC(C = 1, loss='hinge')
svc.fit(x_train, y_train)
svc_pred = svc.predict(x_test)
svc_acc = accuracy_score(svc_pred, y_test)
print("Test Accuraccy : {:.2f}%".format(svc_acc*100))
```

```python
#printing Confusion matrix and classification report
print(confusion_matrix(y_test, svc_pred))
print("\n")
print(classification_report(y_test, svc_pred))
```
